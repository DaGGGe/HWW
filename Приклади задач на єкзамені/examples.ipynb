### Приклад № 1

Написати програмний код, який по введеному коду фірми поверне на екран 
телефон контактної особи фірми

```python
def get_phone():
    """поверне на екран телефон контактної особи фірми
    """
    
    with open('firms.txt', encoding='utf-8') as firms_file:
        firms_list = firms_file.readlines()

#Підготовка массива    
    firms = []
    for line in firms_list:
        firm = line.split(',')
        firms.append(firm)
    
    firm_code = input("Введіть код фірми: ")
    
    for firm in firms:
        if firm_code == firm[0]:
            return firm[2]
            
        
phone_number = get_phone()
if phone_number == None:
    print("Код фірми не існує")
else:
    print(f'Номер фірми = {phone_number}')
```
